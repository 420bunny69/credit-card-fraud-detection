DAY2: feature engineering and smote


```python
from sklearn.preprocessing import StandardScaler
```

first, we would like to scale amount and time, why?-> because other features like v1 to v28 are already scaled & transformed using PCA, and ml algorithms perform poorly when features have vastly different ranges

```python
scaler=StandardScaler()
df["Scaled_Amount"]=scaler.fit_transform(df[["Amount"]])
df["Scaled_Time"]=scaler.fit_transform(df[["Time"]])
df=df.drop(columns=["Amount","Time"])
```

preparing features and target

```python
X=df.drop("Class",axis=1)
#axis=1 is same as columns=["Class"]
y=df["Class"]
print(X.shape)
print("")
print(y.value_counts())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.2, #means use only 20% data for testing
    random_state=42,
    stratify=y #used in imbalanced datasets, so that proportion is preserved in training and testing set
)
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("Train fraud count:", y_train.sum())
print("Test fraud count:", y_test.sum())
```

Applying SMOTE: Synthetic Minority Over-Sampling TEchnique

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=3)

X_train_sm,y_train_sm=smote.fit_resample(X_train,y_train)
print("Before SMOTE:", y_train.value_counts())
print("after SMOTE:", pd.Series(y_train_sm).value_counts())
#why did we convert y_train_sm to Series? -> 
#fit_resample() can return y_train_sm in different formats depending on the input (NumPy array, pandas Series
#by delibrately using pd.Series, we make sure the methods like value_counts will work without error
#basically done for convenience & code consistency
```

```python
print("X_train_sm shape:", X_train_sm.shape)
print("y_train_sm shape:", y_train_sm.shape)
print("Test set unchanged:", y_test.value_counts().to_dict())
```

day 3 training models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
#importing metrics to evaluate the models
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
#importing time to measure training speed
import time

models={
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100,random_state=42),
    "XGBoost": XGBClassifier(random_state=42, eval_metric="logloss")
}
results={} #here we will store results

#using loop to traverse over each model
for name,model in models.items():
    start=time.time()
    model.fit(X_train_sm,y_train_sm)
    duration=time.time()-start

    y_pred=model.predict(X_test)
    y_prob=model.predict_proba(X_test)[:,1] #this line explained in following markdown

    #storing in results
    results[name]={
        "model":model,
        "y_pred":y_pred,
        "AUC":roc_auc_score(y_test,y_prob),
        "Duration":duration
    }
    print("--------------------------------------------------------------------------------")
    print(f"{name} (trained in {duration:.1f}s)")
    print(classification_report(y_test, y_pred, target_names=["Normal", "Fraud"]))
    print(f"AUC-ROC: {roc_auc_score(y_test, y_prob):.4f}")
```

meaning of the line: y_prob = model.predict_proba(X_test)[:,1]

predict_proba() returns: [Probability of class 0,Probability of class 1]
therefore since we only want prob of class 1 (fraud), we do [:,1] which means

"Take every row, but only the second column."

```python
#visualising
fig,axes=plt.subplots(1,3,figsize=(15,4))
for idx,(name,result) in enumerate(results.items()): 
    #idx tells Python which subplot to draw on
    cm=confusion_matrix(y_test,result["y_pred"])
    sns.heatmap(cm, annot=True, fmt="d", ax=axes[idx],
                cmap="Blues",
                xticklabels=["Normal", "Fraud"],
                yticklabels=["Normal", "Fraud"])
    axes[idx].set_title(f"{name}\nAUC: {result['AUC']:.4f}")
    axes[idx].set_xlabel("Predicted")
    axes[idx].set_ylabel("Actual")

plt.tight_layout()
plt.show()
```

```python
#feature importance
xgb_model=results["XGBoost"]["model"]
importance=pd.DataFrame({
    "Features": X.columns,
    "Importance":xgb_model.feature_importances_
}).sort_values(by="Importance", ascending=False).head(15)

plt.figure(figsize=(10,6))
sns.barplot(data=importance, x="Importance", y="Features")
plt.title("Top 15 Important Features (XGBoost)")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.show()
```

saving best model

```python
import joblib
best_model_name=None
highest_AUC=-1
for modelName in results:
    curr_AUC=results[modelName]["AUC"]
    if(curr_AUC>highest_AUC):
        highest_AUC = curr_AUC
        best_model_name = modelName

best_model=results[best_model_name]['model']
print(f"Best model: {best_model_name} with AUC {results[best_model_name]['AUC']:.4f}")
joblib.dump(best_model,"fraud_model.pkl")
joblib.dump(scaler, "scaler.pkl")
print("Model saved")
```
